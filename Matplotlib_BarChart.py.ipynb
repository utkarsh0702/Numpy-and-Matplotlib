```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
company=['GOOGLE','FB','MSFT','AMZN']
revenue=[123,68,92,110]
profit=[50,32,41,10]
```

```python
# taking length of company to make bar as it take both argument as integers
ypos= np.arange(len(company))
ypos
```

```python
plt.bar(ypos,revenue)
```

```python
# Naming x axis bars with company names
plt.xticks(ypos,company)
plt.ylabel('Revenue(bin)')
plt.title('Tech Stocks')
plt.bar(ypos,revenue,label='Revenue')
plt.legend()
```

```python
# ploting two bars
plt.xticks(ypos,company)
plt.ylabel('Revenue(bin)')
plt.title('Tech Stocks')
plt.bar(ypos,revenue,label='Revenue')
plt.bar(ypos,profit,label='Profit')
plt.legend()
```

```python
# to adjust width of two bars
plt.xticks(ypos,company)
plt.ylabel('Revenue(bin)')
plt.title('Tech Stocks')
plt.bar(ypos-0.2,revenue,width=0.4,label='Revenue')
plt.bar(ypos+0.2,profit,width=0.4,label='Profit')
plt.legend()
```

```python
# barh() function to make horizontal bars
plt.yticks(ypos,company)

plt.title('Tech Stocks')
plt.barh(ypos-0.2,revenue,label='Revenue')
plt.barh(ypos+0.2,profit,label='Profit')
plt.legend()
```
