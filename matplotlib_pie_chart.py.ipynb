```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_values=[1400,600,300,410,250]
exp_name=['Home Rent','Food','Phone/Internet Bill','Car','Other Utilities']
```

```python

plt.pie(exp_values,labels=exp_name)
plt.show()
```

```python
# to show axis equally
plt.axis('equal')

plt.pie(exp_values,labels=exp_name)
plt.show()
```

```python
# by default radius of pie is 1 and to change it
plt.axis('equal')
plt.pie(exp_values,labels=exp_name,radius=1.5)
plt.show()
```

```python
# to show percentage of area covered we use autopct argument
plt.axis('equal')
plt.pie(exp_values,labels=exp_name,radius=1.5, autopct='%0.1f%%')
plt.show()
```

```python
# explode argument will take out given peice of pie
plt.axis('equal')
plt.pie(exp_values,labels=exp_name,radius=1.5, autopct='%0.1f%%',explode=[0,0.2,0.1,0,0.1])
plt.show()
```

```python
#startangle will help you to rotate your pie chart to given angle
plt.axis('equal')
plt.pie(exp_values,labels=exp_name,radius=1.5, autopct='%0.1f%%',explode=[0,0.2,0.1,0,0.1],startangle=45)
plt.show()
```
