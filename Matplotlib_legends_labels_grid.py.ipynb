```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
days= [1,2,3,4,5,6,7]
max_t=[50,51,49,52,54,46,50]
min_t=[40,41,44,43,47,41,44]
avg_t=[45,46,47,46,48,43,47]
```

```python
# giving labels to x,y axis
plt.xlabel('Days')
plt.ylabel('Temprature')
plt.title('Weather')

#specifing graph lines
plt.plot(days,max_t,label='Max')
plt.plot(days,min_t,label='Min')
plt.plot(days,avg_t,label='Avg')

#setting legend to best location possible
plt.legend(loc='best')
```
