```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x= [1,2,3,4,5,6,7]
y= [1,21,4,43,5,65,7]
#to plot a graph
plt.plot(x,y)
```

```python
#to seek visual properties
# -- stand for type of lines(dashed lines)
# r stand for colour(red)
# D stand for marker(Diamond)
plt.plot(x,y,'--rD')
```

```python
plt.plot(x,y,color='red',marker='^',linestyle= ':')
```

```python
# #8848A4 is any colour in hexadecimal form
plt.plot(x,y,color='#8848A4',marker='*',linestyle='-.')
```

```python
# regulating marker size
plt.plot(x,y,color='#8848A4',marker='*',linestyle='-.',markersize=20)
```

```python
# alpha() function will control transparency of graph(range is from 0 to 1)
plt.plot(x,y,color='#8848A4',marker='*',linestyle='-.',markersize=20, alpha= 0.2)
```

```python
# plot multiple graphs
x1= [4,6,9,12,15,18,21]
y1= [12,58,37,78,23,56,32]
plt.plot(x,y,'--rD', x1,y1,':g*')

x2=[12,23,34,45,56,67,78]
y2=[34,56,25,57,53,37,12]
plt.plot(x,y,'--rD', x1,y1,':g*',x2,y2,'-.y+')
```
